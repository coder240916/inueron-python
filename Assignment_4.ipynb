**9. What are the list concatenation and list replication operators?**

```python
List concatenation combines two lists and produces a list with elements from both the lists. 

List replication replicates/repeats  all the elements in the list depending on replication number given.

Examples are,

[1,2,3] +[4,5,6]  produces a list [1,2,3,4,5,6]
[1,2,3] * 2 produces [1,2,3,1,2,3]
```

```python
10. What is difference between the list methods append() and insert()?
```

```python
append() method adds a element at the end of the list
insert() method inserts a element at the specified index. 

l = [1,2,3]
l.append(4)  # [1,2,3,4]
l.insert(0,0) # [0,1,2,3,4] places element 0 at index 0.
```

```python
11. What are the two methods for removing items from a list?
```

```python
remove(),pop() methods can be used for removing elements from a list.

l = [1,2,3]
l.remove(1)  # now l =[2,3]
l.pop(1) # removes element from index position 1 and return it. now l =[1,3]
l.pop()  # if no index is given it will remove last element.
```

16. Variables that &quot;contain&quot; list values are not necessarily lists themselves. Instead, what do they
contain?

```python
variables assigned to list values contain references to memory location where list values are stored.
 Using this reference, list values can be accessed and modified.

 For example,

 l1 = [1,2,3]
 l2 = l1

 l2.append(5) # l2 also references same list as l1. values 5 is added to the list.
 print(l1) # [1,2,3,5]  
```

17. How do you distinguish between copy.copy() and copy.deepcopy()?

```python
# copy.copy() creates shallow  copy of a object which is independent from original list .
#  However,If the original list contains any mutable objects such as lists or dictionaries, changes made in copied list are reflected in original list.
# it means copied list contains references of mutable objects in orinal list.

#  For example,
import copy 
original_list = [1,[1,2],3]
copied_list = copy.copy(original_list)


print('-'*25)
print('copy.copy() method')
print('-'*25)
print('original_list ',original_list) 
print('copied_list ',copied_list )
copied_list[1].append(4)
print('original_list after chnages to copied list ', original_list) 
print('copiedl_list after chnages to copied list ', copied_list)


# copy.deepcopy() produces a deep copy of an object.
# If applied to a list, copied list will copy all values from original list including mutable object vlaues.
# chnages made in copied list will not affect original list.

#  For example,
 
original_list = [1,[1,2],3]
copied_list = copy.deepcopy(original_list)

print('-'*25)
print('copy.deepcopy() method')
print('-'*25)
print('original_list ',original_list) 
print('copied_list  ',copied_list )
copied_list[1].append(4)
print('original_list after chnages to copied list ', original_list) 
print('copiedl_list after chnages to copied list ', copied_list)

```
